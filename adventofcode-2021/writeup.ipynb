```python
import numpy as np

TEST = False
DEBUG = False

X = """6,10
0,14
9,10
0,3
10,4
4,11
6,0
6,12
4,1
0,13
10,12
3,4
3,0
8,4
1,10
2,14
8,10
9,0

fold along y=7
fold along x=5"""

if not TEST:
    with open("aoc_13_input.txt", "r") as datafile:
        X = datafile.read()

rawData,rawOps = X.strip().split("\n\n")
dataset = []
for index,line in enumerate(rawData.strip().split("\n")):
    a,b = line.split(",")
    dataset.append( [int(a),int(b)] )


ops = []
for index,op in enumerate(rawOps.strip().split("\n")):
    a,b = op.split("=")
    ops.append( (a[-1], int(b)) )

size=[0,0]
for x,y in dataset:
    if x >= size[0]:
        size[0] = x
    if y >= size[1]:
        size[1] = y
size[0] += 1 if (size[0] % 2)==0 else 2
size[1] += 1 if (size[1] % 2)==0 else 2


def getMap():
    tempMap = np.zeros([size[1],size[0]])
    for x,y in dataset:
        tempMap[y][x] = 1
    return tempMap
```

```python
currentMap = getMap()

for t,index in ops[:1]:
    currentMap = getMap()
    #print(currentMap)
    print(size)
    print(t, index)
    print()
    FLIPPED = np.flip(currentMap, 0) if t == "y" else np.flip(currentMap, 1)
    
    if t == "y":
        size[1] = index
    else:
        size[0] = index

    markers = []
    
    for y,x in np.argwhere( currentMap == 1 ):
        if x < size[0] and y < size[1]:
            markers.append( (x,y) )
            
    for y,x in np.argwhere( FLIPPED == 1 ):
        if x < size[0] and y < size[1]:
            markers.append( (x,y) )
    
    dataset = list(set(markers))

print(size)
print(getMap())

print( len( np.argwhere( currentMap == 1 ) ) )
```
