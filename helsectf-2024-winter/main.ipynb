## reisetid - 100
Kari og Ola bor 150km fra hverandre og skal møtes for å ta en kaffe.  
Kari kjører kl 1200, og ville brukt 3 timer om hun måtte kjøre hele lengden på 150km.  
Ola kjører kl 1300, og ville brukt 5 timer om han måtte kjøre hele lengden på 150km.

Hva er klokka når de møtes, og hvor langt i meter har Kari kjørt?

Flaggformat: `helsectf{<tid 4tall>_<lengde i meter>}`

### Answer
See code below  
**Flag:** `helsectf{1415_112500}`

````python
dist = 150
kari_dep = 1200
kari_time = 3
kari_speed = dist / kari_time
ola_dep = 1300
ola_time = 5
ola_speed = dist/ola_time
kari_headstart = kari_speed * 1
for minutes in range(75,100,1):
    hours = minutes / 60
    ola_dist = ola_speed * hours
    kari_dist = kari_speed * hours
    if kari_headstart+ola_dist+kari_dist == 150:
        print("helsectf{"+f"{12+((60+minutes)//60)}{(75+60)%60}_{int((kari_headstart+kari_dist)*1000)}"+"}")
````

## image processing 1 - 100
Fourier-transformasjoner kan brukes til mye innenfor bildeprosessering. man kan lett beregne en diskret fourier-transformasjon til et bilde i python:  
`ft = numpy.fft.fft2(img_data)`  
en vanlig måte å visualisere en kompleks fourier transformasjon på er å skyve den slik at origo, og de laveste frekvensene havner i midten av visualiseringen.  
For å visualisere komplekse tall i vanlige grå-farger tar man gjerne normen / absoluttverdien til de komplekse verdiene.  
Det er også vanlig å visualisere det med en logaritmisk skala, siden endringene i frekvens-intensiteten ofte er numerisk liten:

```
fshift = numpy.fft.fftshift(ft)
spectrum = numpy.log(numpy.abs(fshift))
```  
hva skjuler seg i dette bildet av et ekorn?

<img src="./data/ekorn.png" width=300px /> <br>

### Answer
**Flag:** `helsectf{krokodille}`

<img src="./data/ekorn.spectrum.png" width=500px />

````python
from PIL import Image
import numpy
img_data = Image.open("./data/ekorn.png")
ft = numpy.fft.fft2(img_data)
fshift = numpy.fft.fftshift(ft)
spectrum = numpy.log(numpy.abs(fshift))
for y in range(len(spectrum)):
    for x in range(len(spectrum[y])):
        if int(spectrum[y][x]) == 9:
            spectrum[y][x] = 128
img = Image.fromarray(spectrum)
img = img.rotate(180)
img.show()
````

## image processing 2 - 269
fourier-transformasjonen til et bilde kan brukes til mye nyttig, blant annet til edge-detection.  
et high-pass filter vil slippe gjennom høye frekvenser, men stoppe lave frekvenser.   
de lave frekvensene beskriver de store og grove detaljene i bildet, mens de høye frekvensene beskriver små detaljer og harde linjer, teksturer og hår f.eks.  
ved å fjerne de lave frekvensene kan vi lettere se de høye frekvensene og kanter og detaljer kommer bedre fram.  
jeg har lagt ved et bilde av et blåbær, men det er dessverre nokså skadet og ting har glidd litt sammen, så det er umulig å se detaljene i bildet.   
se om du klarer å få frem noen detaljer!

tips! prøv gjerne litt forskjellige masker med forskjellige verdier.


<img src="./data/blabr.png" width=400px /> <br>


### Answer
Highpass filter with radius=8  
**Flag:** `helsectf{pepperoni}`

<img src="./data/blabr.hp_filter.png" width=500px />


````python
import numpy as np
import cv2

# read input and convert to grayscale
img = cv2.imread('./data/blabr.png')

# do dft saving as complex output
dft = np.fft.fft2(img, axes=(0,1))

# apply shift of origin to center of image
dft_shift = np.fft.fftshift(dft)

# create white circle mask on black background and invert so black circle on white background
radius = 7
mask = np.zeros_like(img)
cy = mask.shape[0] // 2
cx = mask.shape[1] // 2
cv2.circle(mask, (cx,cy), radius, (255,255,255), -1)[0]
mask = 255 - mask

# blur the mask
mask2 = cv2.GaussianBlur(mask, (19,19), 0)

# apply mask to dft_shift
dft_shift_masked = np.multiply(dft_shift,mask) / 255
dft_shift_masked2 = np.multiply(dft_shift,mask2) / 255

# shift origin from center to upper left corner
back_ishift = np.fft.ifftshift(dft_shift)
back_ishift_masked = np.fft.ifftshift(dft_shift_masked)
back_ishift_masked2 = np.fft.ifftshift(dft_shift_masked2)


# do idft saving as complex output
img_back = np.fft.ifft2(back_ishift, axes=(0,1))
img_filtered = np.fft.ifft2(back_ishift_masked, axes=(0,1))
img_filtered2 = np.fft.ifft2(back_ishift_masked2, axes=(0,1))

# combine complex real and imaginary components to form (the magnitude for) the original image again
# multiply by 3 to increase brightness
img_back = np.abs(img_back).clip(0,255).astype(np.uint8)
img_filtered = np.abs(3*img_filtered).clip(0,255).astype(np.uint8)
img_filtered2 = np.abs(3*img_filtered2).clip(0,255).astype(np.uint8)

# write result to disk
cv2.imwrite("lena_dft_numpy_highpass_filtered2.png", img_filtered2)
````

## Pass or fail?
`.bat` - filer kan gjøre ganske mye.  
Denne bat-filen vil ikke kjøre noe skummelt om du kjører den.  
Finn ut hva den gjør.  
Evt hva du kan mate den med (input) for å få ønsket resultat.

Flagget gjemmer seg i en pass, ikke en fail.

Passord for zip-fil er standard (`infected`).

**Answer**

````python
import re
re_variable = re.compile(rb"%([^:%]+):~(\d+),(\d+)%")
variables = {
    #b"": b"",
    b"fortressVeered": b"C:\\Users\\admin\\AppData\\Local\\Temp'",
    b"turtles": b"respondents\\\\crossbar.%3",
    b"vigilantMutants": b"replace",
    b"fluffyBunny": b"run",
    b"temp": b"C:\\Users\\admin\\AppData\\Local\\Temp'",
    b"starlightPlacebos": b"C:\\Windows'",
    b"windir": b"C:\\Windows'",
    b"flightorfight": b"exe",
    b"counselWaltzer": b"system",
    b"yawsAbolitionists": b"star",
    b"besiegingWaterproofing":b"naststargysystemx"
}

with open("./pass_or_fail/pass_or_fail.bat", "rb") as batfh:
    bat_file = batfh.read().strip()

for variable, index, length in re_variable.findall(bat_file):
    if variable in variables:
        #print(b"%b:~%b,%b      %b" % (variable, index, length, variables[variable][int(index):int(index)+int(length)]))
        bat_file = bat_file.replace(b"%"+b"%b:~%b,%b"%(variable, index, length)+b"%", variables[variable][int(index):int(index)+int(length)])

for key in variables:
    if b"%"+key+b"%" in bat_file:
        bat_file = bat_file.replace(b"%"+key+b"%", variables[key])

print()
for line in bat_file.split(b"\n"):
    if not line.strip():
        print()
    else:
        print(line.strip())
````

## prikker - 400
En TV skjerm har klikket helt og viser bare prikker i ulike farger.  
Hvis man ser godt etter kan man kanskje se en hemmelig melding, spesielt hvis man finner den riktige fargen!

### Answer
After a lot of programming mishaps for extracting the values, and then efficiently extract them I wrote the script below.  

There are 4000 frames in the image, and the initial solution gave the flag in the "R" channel of the image. But it is cluttered. After viewing the summed value distribution of the 4k frames, and setting an arbitrary max sum limit filter of 20k, the following image appeared with the flag.

**Answer:** `helsectf{en_prikk_kan_ha_mange_farger!}`

<img src="./data/prikk.gif" width="800px" />

````python
from PIL import Image
import numpy as np

RGB = np.zeros(shape=(214, 2440), dtype=int)
sum_frames = []
with Image.open("./data/prikker.gif") as img:
    frames = img.n_frames
    for frame in range(frames):
        img.seek(frame)
        frame_image = np.array(img)
        a = Image.fromarray(frame_image)
        bands = a.getbands()
        if "L" in bands:
            continue
        X,Y = a.size
        channel = np.array(a.getchannel("R"))
        sum_frames.append(channel.sum())
        if channel.sum() < 20000:
            RGB += channel
Image.fromarray(RGB).save("prikk.gif")
````
