### Dataforskning
CSV file with simple scatter plot  
converting to ASCII art shows the data easily
**flag:** `PST{EGGESKALL}`

````python
#Dataforskning
import numpy as np

with open("innsamlet.csv","r") as csv:
    data = [row.split(",") for row in csv.read().strip().split("\n")]
    
for cols in data[1:]:
    cols[0] = int(cols[0])
    cols[1] = int(cols[1])
    cols[2] = int(cols[2].replace("'",""))

txt = []
for row in data[1:]:
    txt.append("#" if row[2]==1 else " ")
    if row[0] == 56:
        print(" ".join(txt))
        txt = []
````
