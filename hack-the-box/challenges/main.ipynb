```python
import base64

filepath=r"C:\Users\JollyJackson\Downloads\APKey\APKey\original\META-INF\JOHN.SF"
def_text = "SHA-256-Digest: "
with open(file=filepath) as fh:
    lines = [line[len(def_text):].strip() for line in fh.readlines() if line.startswith(def_text)]

#print(lines)
for line in lines:
    try:
        decoded = base64.b64decode(line)
        print(f"'{line}' == {decoded.decode('utf-8')}")
    except Exception as e:
        print(f"'{line}' - {e}")
        pass
```
